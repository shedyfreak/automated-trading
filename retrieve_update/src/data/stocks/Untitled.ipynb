```python
from bs4 import BeautifulSoup
import requests
article_headers=[]
html_doc=requests.get("https://www.swissinfo.ch/eng/business")
soup = BeautifulSoup(html_doc.text, 'html.parser')
#print(soup.prettify())
w=soup.find_all("p",class_="si-teaser__lead")
s=soup.find_all("a",class_="")
for i in w:
    alpha=i.get_text()
    article_headers.append(alpha.split("\n")[4])

for i in article_headers:
    print(i+"\n")

```
